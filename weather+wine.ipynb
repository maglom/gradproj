```python
from weather import get_weather
import pandas as pd

df = pd.read_csv('scrape/bordeaux/bordeaux.csv')
columns = get_weather('bordeaux', 2000).iloc[1,:]
for c,i in enumerate(columns):
    columns[c] = str(c)+i
df[columns] = 0
```

```python
for index, row in df.iterrows():
    region = 'Bordeaux'
    year = row['Year']
    we = get_weather(region, year)
    df.iloc[index,4:] = we.iloc[2,:]
    print(df.iloc[index,:])
    if index == 10:
        break
```
